```python
import species


def main():
    target = species.StringDNASpecie(DNA='It was the best of times it was the worst of times')

    pop = species.Population(population_size=8_000, target=target)

    i = 0
    while pop.best_fit() < 1.0:
        pop.mutate()
        pop.selection()
        pop.best_fit()
        i += 1
        print(i)

if __name__ == '__main__':
    main()
```
